```python
import numpy as np 
import matplotlib.pyplot as plt 
import json 

path = '/home/deeplearning/Deep_Learning_result/classic_model_test/history.json'
#dataset = []
with open(path) as json_file : 
    history = json.load(json_file)
    #for info in reader :
        #dataset.append(info)
```

```python
history.keys()
```

```python
loss = []
val_loss = []

head_acc = []
head_val_acc = []
leg_acc = []
leg_val_acc = []
right_acc = []
right_val_acc = []
left_acc =  []
left_val_acc = []
for i in range(len(history['loss'])):
    loss.append(float(history['loss'][str(i)]))
    val_loss.append(float(history['val_loss'][str(i)]))
    head_acc.append(float(history['head_accuracy'][str(i)]))
    head_val_acc.append(float(history['val_head_accuracy'][str(i)]))
    leg_acc.append(float(history['leg_accuracy'][str(i)]))
    leg_val_acc.append(float(history['val_leg_accuracy'][str(i)]))
    right_acc.append(float(history['right_arm_accuracy'][str(i)]))
    right_val_acc.append(float(history['val_right_arm_accuracy'][str(i)]))
    left_acc.append(float(history['left_arm_accuracy'][str(i)]))
    left_val_acc.append(float(history['val_left_arm_accuracy'][str(i)]))
```

```python
f = plt.figure(figsize=(8,5))
axes = plt.gca()
plt.plot(np.linspace(0,19,20),loss, label='train', linewidth = 3)
plt.plot(np.linspace(0,19,20),val_loss, label='val', linewidth = 3)
plt.legend()
plt.grid()
plt.xlabel('epochs')
axes.xaxis.set_ticks(range(20))
axes.set_xlim(-1,20)
plt.title('LOSS')

f = plt.figure(figsize=(8,5))
axes = plt.gca()
plt.plot(np.linspace(0,19,20),head_acc, label='train', linewidth = 3)
plt.plot(np.linspace(0,19,20),head_val_acc, label='val', linewidth = 3)
plt.legend()
plt.grid()
plt.xlabel('epochs')
axes.xaxis.set_ticks(range(20))
axes.set_xlim(-1,20)
plt.title('HEAD ACCURACY')

f = plt.figure(figsize=(8,5))
axes = plt.gca()
plt.plot(np.linspace(0,19,20),leg_acc, label='train', linewidth = 3)
plt.plot(np.linspace(0,19,20),leg_val_acc, label='val', linewidth = 3)
plt.legend()
plt.grid()
plt.xlabel('epochs')
axes.xaxis.set_ticks(range(20))
axes.set_xlim(-1,20)
plt.title('LEG ACCURACY')

f = plt.figure(figsize=(8,5))
axes = plt.gca()
plt.plot(np.linspace(0,19,20),right_acc, label='train', linewidth = 3)
plt.plot(np.linspace(0,19,20),right_val_acc, label='val', linewidth = 3)
plt.legend()
plt.grid()
plt.xlabel('epochs')
axes.xaxis.set_ticks(range(20))
axes.set_xlim(-1,20)
plt.title('RIGHT ARM ACCURACY')

f = plt.figure(figsize=(8,5))
axes = plt.gca()
plt.plot(np.linspace(0,19,20),left_acc, label='train', linewidth = 3)
plt.plot(np.linspace(0,19,20),left_val_acc, label='val', linewidth = 3)
plt.legend()
plt.grid()
plt.xlabel('epochs')
axes.xaxis.set_ticks(range(20))
axes.set_xlim(-1,20)
plt.title('LEFT ARM ACCURACY')
```
